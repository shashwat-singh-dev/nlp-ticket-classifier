```python
import re

def clean_text(text):
    text = text.lower()  # sab lowercase
    text = re.sub(r'[^a-zA-Z ]', '', text)  # special characters remove
    return text

df["clean_text"] = df["ticket_text"].apply(clean_text)

df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(df["clean_text"])
```

```python
X.shape
```

```python
vectorizer.get_feature_names_out()
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(X, df["category"])
```

```python
model
```

```python
test = ["payment not working"]

# clean karo
test_clean = [clean_text(t) for t in test]

# numbers me convert
test_vec = vectorizer.transform(test_clean)

# prediction
model.predict(test_vec)
```

```python
def get_priority(text):
    text = text.lower()

    if "urgent" in text or "hacked" in text or "breach" in text:
        return "High"
    elif "failed" in text or "not working" in text or "error" in text:
        return "Medium"
    else:
        return "Low"
```

```python
def predict_ticket(text):
    clean = clean_text(text)
    vec = vectorizer.transform([clean])

    category = model.predict(vec)[0]
    priority = get_priority(text)

    return category, priority
```
